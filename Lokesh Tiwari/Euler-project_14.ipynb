#### the function named 'operate' return number of number in collatz sequence of a number num given as parameter

```python
import numpy as np
from numba import njit
import time
```

```python
@njit
def operate(num):
    k = num
    count = 1
    while True:
        if k == 1:
            return count
        elif k%2==0:
            count += 1
            k=int(k/2)
        else:
            count += 1
            k=3*k+1
```

```python
operate(13)
```

```python
def longestChainLengthBelow(num):
    ans = 1
    chain_length = 1
    for i in range(1, num):
        z = operate(i)
        if z > chain_length:
            chain_length = z
            ans = i
    print(chain_length, ' is longest chain length for number ', ans)
```

```python
longestChainLengthBelow(1000000)
```
