```python
from numba import jit
import time
import math as m
@jit(nopython=True)
def tellprime(z):
    if z==2:
        return True
    elif z==3:
        return True
    else:
        i = 2
        l = m.sqrt(z)
        u = True 
        while i<=l and u:
            if z%i==0:
                return False
                break
            elif i+1>l and z%i != 0:
                return True
                break
            else:
                i+=1
            u = (i%6 in [1,5]) or (i in [2,3])
            while u == False:
                i+=1
                if z%i==0:
                    return False
                    break
                elif i+1>l and z%i != 0:
                    return True
                    break
                u = i%6 in [1,5]
```

```python
#The 570,004,517,329th prime is 16,765,962,748,063.                
start = time.time()               
print(tellprime(16765962748063))
end = time.time()
print(end-start)
```

```python
def operate(n):
    y = list(str(n))
    if len(y)==len(set(y)):
        del y
        if tellprime(n):
            return True
        else:
            return False
    else:
        return False
```
