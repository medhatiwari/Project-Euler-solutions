```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def operate(d):
    l1 = []
    n = 1
    while True:
        n = ((n*10)%d)
        if n == 0:
            return 0
        elif n in l1:
            return len(l1)
        l1.append(n)
```

```python
operate(2)
```

```python
d = np.arange(1,1000) 
```

```python
recurring_cycle = []
```

```python
for i in d:
    recurring_cycle += [operate(i)]
```

```python
recurring_cycle.index(max(recurring_cycle))
```

```python
max(recurring_cycle)
```

```python
operate(983)
```

```python
x_axis = np.array(d[0:500])
y_axis = np.array(recurring_cycle[0:500])
```

```python
plt.plot(x_axis,y_axis,'r+',scalex=True,scaley=True)
```

```python
plt.scatter(x_axis,y_axis,s=5)
```
